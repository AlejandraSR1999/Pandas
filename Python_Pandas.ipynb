Groupby

```python
df_aux = df.groupby(['continent', 'location']).count()[['new_deaths_smoothed']].rename(columns={"new_deaths_smoothed":"new_deaths_sthd"})
```

```python
df_aux
```

Sorting

```python
df_aux_sorted = df_aux.sort_values("new_deaths_sthd", ascending=False)
df_aux_sorted
```

Reset_Index

```python
df_aux_sorted = df_aux_sorted.reset_index()
```

```python
df_aux_sorted['continent'].nunique()
```

```python
df_aux_sorted['location'].nunique()
```

Pivot Table

```python
df_aux_sorted.head()
```

```python
df_final = df_aux_sorted.pivot_table(values='new_deaths_sthd', index='location', columns='continent').fillna("")
df_final
```

Merge

```python
df_A = df_final.reset_index()
```

```python
df_B = pd.DataFrame({"location":["Zimbabwe", 'Zambia', 'Albania', 'NewCountry'], "red_flag":["Yes", "Yes", "TBD", "NOOOO"]})
```

```python
df_B
```

```python
df_A.head()
```

```python
df_merged = pd.merge(df_A, df_B, how='outer', left_on='location', right_on='location', indicator=True)
```

```python
df_merged
```

```python
df_merged.groupby('_merge').count()
```
